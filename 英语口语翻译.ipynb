```python
from math import sqrt 
def fibol(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibol(n - 1) + fibol(n - 2)

def fibo2(n):
    return ((1 + sqrt(5))**n-(1-sqrt(5))**n)/(2**n*sqrt(5))
%timeit fibol(1)
```
